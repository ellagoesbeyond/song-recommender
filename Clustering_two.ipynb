```python
scaler2 = StandardScaler()
scaler2.fit(sn_db)
sn_db_scaled = scaler2.transform(sn_db)
sn_db_scaled_df = pd.DataFrame(sn_db_scaled, columns = sn_db.columns)
display(sn_db.head())
print()
display(sn_db_scaled_df.head())
```

```python
sn_db_scaled_df.describe()
```

```python
with open("Model/scaler2.pickle", "wb") as f:
    pickle.dump(scaler2,f)
    
```

```python
K = range(2, 21)
inertia = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(sn_db)
    inertia.append(kmeans.inertia_)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
kmeans = KMeans(n_clusters=11, random_state=1234)
kmeans.fit(sn_db_scaled_df)
```

```python
# assign a cluster to each example
labels = kmeans.predict(sn_db_scaled_df)
# retrieve unique clusters
clusters = np.unique(labels)
# create scatter plot for samples from each cluster
for cluster in clusters:
    # get row indexes for samples with this cluster
    row_ix = np.where(labels == cluster)
    # create scatter of these samples
    pyplot.scatter(sn_db.to_numpy()[row_ix, 1], sn_db.to_numpy()[row_ix, 3])
    # show the plot
pyplot.show()
```

```python
clusters = kmeans.predict(sn_db_scaled_df)
#clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
sn_db["cluster"] = clusters
sn_db
```
